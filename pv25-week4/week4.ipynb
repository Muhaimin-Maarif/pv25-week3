```python
import sys
from PyQt6.QtWidgets import (QApplication, QWidget, QVBoxLayout, QComboBox, QFormLayout, QLineEdit, QPushButton,
                             QHBoxLayout, QTextEdit, QLabel
                             )

class Task4(QWidget):
    def __init__(self):
        super().__init__()
        self.resize(400,400)
        self.setWindowTitle("Cashier App")

        self.daftar_product = {
            "Beras": 20000,
            "Minyak": 25000,
            "Kecap": 5000,
            "Saos": 2500
        }

        self.daftar_discount = {
            "5%": 5,
            "10%": 10,
            "15%": 15
        }

        self.iniUI()

        self.tampilan = []
        self.total = 0
        
    def iniUI(self):
        layout = QVBoxLayout()

        form = QFormLayout()

        self.product = QComboBox()
        self.product.addItems(self.daftar_product.keys())
        form.addRow("Product", self.product)

        self.quantity = QLineEdit()
        form.addRow("Quantity", self.quantity)

        self.discount = QComboBox()
        self.discount.addItem("0%")
        self.discount.addItems(self.daftar_discount.keys())
        form.addRow("Discount", self.discount)

        btn = QHBoxLayout()
        self.btn_add = QPushButton("Add to Cart")
        self.btn_clear = QPushButton("Clear")
        btn.addWidget(self.btn_add)
        btn.addWidget(self.btn_clear)
        self.btn_add.clicked.connect(self.btnAdd)
        self.btn_clear.clicked.connect(self.btnClear)

        self.tampilan_text = QTextEdit()
        self.tampilan_text.setReadOnly(True)

        self.total_text = QLabel("Total: Rp. 0")

        layout.addLayout(form)
        layout.addLayout(btn)
        layout.addWidget(self.tampilan_text)
        layout.addWidget(self.total_text)
        layout.addWidget(QLabel("Created By Muhaimin"))
        self.setLayout(layout)

    def btnAdd(self):
        nama_product = self.product.currentText()
        value_product = self.daftar_product[nama_product]

        try:
            value_quantity = int(self.quantity.text())
        except ValueError:
            self.tampilan_text.setText("Quantity Harus Angka")
            return

        nama_discount = self.discount.currentText()
        value_discount = self.daftar_discount.get(nama_discount, 0)
        total = (value_product * value_quantity) - ((value_product * value_quantity) * value_discount / 100)

        text = f'{nama_product} (Rp. {value_product}) - {value_quantity} x Rp. {value_product}'
        if value_discount > 0 and nama_discount.strip() != "":
            text += f' (disc {nama_discount})'
        text += f' = {total} \n'
        self.tampilan.append(text)

        self.total += total

        self.update()

    def update(self):
        self.tampilan_text.setText("".join(self.tampilan))
        self.total_text.setText(f'Total: Rp. {self.total}')

    def btnClear(self):
        self.tampilan = []
        self.total = 0
        self.tampilan_text.clear()
        self.total_text = "Total: Rp. 0"

if __name__ == "__main__":
    app = QApplication(sys.argv)
    window = Task4()
    window.show()
    sys.exit(app.exec())
```
